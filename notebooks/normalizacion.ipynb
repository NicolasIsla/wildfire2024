```python
import os
import json
import re
import numpy as np

def explorador_2019a(path_images, path_labels):
    images = os.listdir(path_images)
    labels = os.listdir(path_labels)
    print(f"Images: {len(images)} Labels: {len(labels)}")
    print()
    
    cat = [re.split(r'[_.]', img)[0] for img in images]
    
    # Contar las categorías
    unique, counts = np.unique(cat, return_counts=True)
    print(f"Videos: {len(unique)}")
    print()
    # Mostrar las categorías
    for i in range(len(unique)):
        print(f"{unique[i]}: {counts[i]}")
    
    return unique

def generate_json(path_images, path_labels, output_file):
    data_train = {
        'images': [],
        'annotations': [],
        'categories': []
    }

    # Assuming categories are predefined
    categories = [{'id': 1, 'name': 'smoke'}, {'id': 0, 'name': 'nothing'}]
    data_train['categories'] = categories

    unique_videos = explorador_2019a(path_images, path_labels)
    video_mapping = {video: idx for idx, video in enumerate(unique_videos)}

    image_id = 0
    annotation_id = 0

    image_files = sorted([f for f in os.listdir(path_images) if f.endswith('.jpg') or f.endswith('.png')])
    
    frame_counters = {video: 0 for video in unique_videos}

    for image_file in image_files:
        image_path = os.path.join(path_images, image_file)
        label_file = image_file.replace('.jpg', '.txt').replace('.png', '.txt')
        label_path = os.path.join(path_labels, label_file)

        video_id_str = re.split(r'[_.]', image_file)[0]
        video_id = video_mapping.get(video_id_str)
        
        # Asignar el frame_id basado en el contador para cada video_id
        frame_id = frame_counters[video_id_str]
        frame_counters[video_id_str] += 1

        # Reading image info
        image_info = {
            'id': image_id,
            'file_name': image_file,
            'path': image_path,
            'id_video': video_id,
            'id_frame': frame_id
        }
        data_train['images'].append(image_info)

        # Reading annotation info
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                annotations_exist = False
                for line in f:
                    parts = line.strip().split()
                    category_id = int(parts[0])
                    bbox = list(map(float, parts[1:]))
                    annotation_info = {
                        'id': annotation_id,
                        'image_id': image_id,
                        'category_id': 1,
                        'bbox': bbox,
                        'id_video': video_id,
                        'id_frame': frame_id
                    }
                    data_train['annotations'].append(annotation_info)
                    annotations_exist = True
                    annotation_id += 1
                if not annotations_exist:
                    annotation_info = {
                        'id': annotation_id,
                        'image_id': image_id,
                        'category_id': 0,
                        'bbox': [],
                        'id_video': video_id,
                        'id_frame': frame_id
                    }
                    data_train['annotations'].append(annotation_info)
                    annotation_id += 1
        else:
            annotation_info = {
                'id': annotation_id,
                'image_id': image_id,
                'category_id': 0,
                'bbox': [],
                'id_video': video_id,
                'id_frame': frame_id
            }
            data_train['annotations'].append(annotation_info)
            annotation_id += 1

        image_id += 1



    return data_train

path_images = r'D:\dataincendios\2019a-smoke-full\DS\images\train'
path_labels = r'D:\dataincendios\2019a-smoke-full\DS\labels\train'
output_file = 'data_train.json'

data_train = generate_json(path_images, path_labels, output_file)

print(data_train['images'][0])
print(data_train['annotations'][0])
print(data_train['categories'][0])
```

## SmokeFrames-2.4k

```python
import os
import json
import re
import numpy as np


def explorador_smoke(path_images, path_labels):
    images = os.listdir(path_images)
    labels = os.listdir(path_labels)
    print(f"Images: {len(images)} Labels: {len(labels)}")
    print()
    
    # cat = [re.split(r'[_.]', img)[0] for img in images]
    # cat ahora sea dividido por la palabra frame
    cat = [re.split(r'_frame', img)[0] for img in images]

    
    # Contar las categorías
    unique, counts = np.unique(cat, return_counts=True)
    print(f"Videos: {len(unique)}")
    print()
    # Mostrar las categorías
    for i in range(len(unique)):
        print(f"{unique[i]}: {counts[i]}")
    
    return unique

def generate_json(path_images, path_labels, output_file):
    data_train = {
        'images': [],
        'annotations': [],
        'categories': []
    }

    # Assuming categories are predefined
    categories = [{'id': 1, 'name': 'smoke'}, {'id': 0, 'name': 'nothing'}]
    data_train['categories'] = categories

    unique_videos = explorador_smoke(path_images, path_labels)
    video_mapping = {video: idx for idx, video in enumerate(unique_videos)}

    image_id = 0
    annotation_id = 0

    image_files = sorted([f for f in os.listdir(path_images) if f.endswith('.jpg') or f.endswith('.png')])
    
    frame_counters = {video: 0 for video in unique_videos}

    for image_file in image_files:
        image_path = os.path.join(path_images, image_file)
        label_file = image_file.replace('.jpg', '.txt').replace('.png', '.txt')
        label_path = os.path.join(path_labels, label_file)

        # Extract video_id and frame_id from the file name
        match = re.match(r'vid_(\d+)_frame_(\d+)\.jpg', image_file)
        if match:
            video_id_str = f"vid_{match.group(1)}"
            frame_id = int(match.group(2))
            video_id = video_mapping.get(video_id_str)
        else:
            print(f"Filename '{image_file}' does not match the expected format.")
            continue

        # Reading image info
        image_info = {
            'id': image_id,
            'file_name': image_file,
            'path': image_path,
            'id_video': video_id,
            'id_frame': frame_id
        }
        data_train['images'].append(image_info)

        # Reading annotation info
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                annotations_exist = False
                for line in f:
                    parts = line.strip().split()
                    category_id = int(parts[0])
                    bbox = list(map(float, parts[1:]))
                    annotation_info = {
                        'id': annotation_id,
                        'image_id': image_id,
                        'category_id': 1,
                        'bbox': bbox,
                        'id_video': video_id,
                        'id_frame': frame_id
                    }
                    data_train['annotations'].append(annotation_info)
                    annotations_exist = True
                    annotation_id += 1
                if not annotations_exist:
                    annotation_info = {
                        'id': annotation_id,
                        'image_id': image_id,
                        'category_id': 0,
                        'bbox': [],
                        'id_video': video_id,
                        'id_frame': frame_id
                    }
                    data_train['annotations'].append(annotation_info)
                    annotation_id += 1
        else:
            annotation_info = {
                'id': annotation_id,
                'image_id': image_id,
                'category_id': 0,
                'bbox': [],
                'id_video': video_id,
                'id_frame': frame_id
            }
            data_train['annotations'].append(annotation_info)
            annotation_id += 1

        image_id += 1

    # Guardar el JSON en un archivo
    with open(output_file, 'w') as outfile:
        json.dump(data_train, outfile, indent=4)

    return data_train


path_images = r'D:\dataincendios\SmokeFrames-2.4k\DS\images\train'

path_labels = r'D:\dataincendios\SmokeFrames-2.4k\DS\labels\train'

output_file = 'data_train.json'

data_train = generate_json(path_images, path_labels, output_file)

print(data_train['images'][0])
print(data_train['annotations'][0])
print(data_train['categories'][0])
```
