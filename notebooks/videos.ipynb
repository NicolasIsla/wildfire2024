```python
import os
import shutil
import random

path_destino_train_images = r"C:\nico\wildfire2024\data\PyroVideos\images\train"
path_destino_train_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\train"
path_destino_test_images = r"C:\nico\wildfire2024\data\PyroVideos\images\test"
path_destino_test_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\test"
path_destino_val_images = r"C:\nico\wildfire2024\data\PyroVideos\images\val"
path_destino_val_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\val"

os.makedirs(path_destino_train_images, exist_ok=True)
os.makedirs(path_destino_train_labels, exist_ok=True)
os.makedirs(path_destino_test_images, exist_ok=True)
os.makedirs(path_destino_test_labels, exist_ok=True)
os.makedirs(path_destino_val_images, exist_ok=True)
os.makedirs(path_destino_val_labels, exist_ok=True)

split_train = 0.8
split_val = 0.1
split_test = 0.1

incendios_videos = os.listdir(path_figlib)
print(len(incendios_videos))

random.shuffle(incendios_videos)
n_train = int(len(incendios_videos) * split_train)
n_val = int(len(incendios_videos) * split_val)
n_test = len(incendios_videos) - n_train - n_val

# Inicializamos los contadores
total_train_images = 0
total_val_images = 0
total_test_images = 0

train_incendios = 0
val_incendios = 0
test_incendios = 0

# Función para copiar archivos a la carpeta correspondiente
def copiar_frames_y_labels(frames, labels, destino_images, destino_labels, video_name):
    for i, frame in enumerate(frames):
        frame_dest = os.path.join(destino_images, f"{video_name}_frame{i}.jpg")
        label_dest = os.path.join(destino_labels, f"{video_name}_frame{i}.txt")
        
        # Copiar frame y etiqueta
        shutil.copy(frame, frame_dest)
        if os.path.exists(labels[i]):

            shutil.copy(labels[i], label_dest)
        else:
            # Si no existe la etiqueta, crear un archivo vacío
            with open(label_dest, 'w') as f:
                pass

# Asignar incendios al conjunto de entrenamiento, validación y prueba
for idx, incendio in enumerate(incendios_videos):
    path_video = os.path.join(path_figlib, incendio)
    
    # Obtener los frames y labels
    frames = [os.path.join(path_video, f) for f in os.listdir(path_video) if f.endswith('.jpg')]
    labels_path = os.path.join(path_video, 'labels')  # Subcarpeta donde están los labels
    labels = [os.path.join(labels_path, f) for f in os.listdir(labels_path) if f.endswith('.txt')]
    
    # Ordenar alfabéticamente por nombre
    frames.sort()
    labels.sort()

    video_name = incendio

    # Clasificar en train, val o test según el índice
    if idx < n_train:
        copiar_frames_y_labels(frames, labels, path_destino_train_images, path_destino_train_labels, video_name)
        total_train_images += len(frames)
        train_incendios += 1
    elif idx < n_train + n_val:
        copiar_frames_y_labels(frames, labels, path_destino_val_images, path_destino_val_labels, video_name)
        total_val_images += len(frames)
        val_incendios += 1
    else:
        copiar_frames_y_labels(frames, labels, path_destino_test_images, path_destino_test_labels, video_name)
        total_test_images += len(frames)
        test_incendios += 1

# Imprimir resultados
print("Separación completa.")
print(f"Total imágenes entrenamiento: {total_train_images}")
print(f"Total imágenes validación: {total_val_images}")
print(f"Total imágenes prueba: {total_test_images}")
print(f"Total incendios en entrenamiento: {train_incendios}")
print(f"Total incendios en validación: {val_incendios}")
print(f"Total incendios en prueba: {test_incendios}")

```

```python
import os
import re
import random
import shutil
from collections import defaultdict

# Directorio base
path_pyro = r"C:\Users\corpa\Downloads\pyronear_ds_03_2024\pyronear_ds_03_2024"

# Rutas de etiquetas e imágenes
path_labels_train = os.path.join(path_pyro, "labels", "train")
path_labels_val = os.path.join(path_pyro, "labels", "val")
path_images_train = os.path.join(path_pyro, "images", "train")
path_images_val = os.path.join(path_pyro, "images", "val")

# Obtener lista de archivos con paths completos
cantidad_labels_train = [os.path.join(path_labels_train, f) for f in os.listdir(path_labels_train)]
cantidad_labels_val = [os.path.join(path_labels_val, f) for f in os.listdir(path_labels_val)]
cantidad_images_train = [os.path.join(path_images_train, f) for f in os.listdir(path_images_train)]
cantidad_images_val = [os.path.join(path_images_val, f) for f in os.listdir(path_images_val)]

# Combinar todas las imágenes y etiquetas
cantidad = cantidad_labels_train + cantidad_labels_val + cantidad_images_train + cantidad_images_val

# Términos de búsqueda
terminos = ["AWF", "hpwren", "pyronear", "random", "synthetic", "ADF"]

# Inicializar diccionario para agrupaciones
agrupaciones = {termino: [] for termino in terminos}
agrupaciones["otros"] = []  # Para los elementos que no coincidan con ningún término

# Agrupar elementos según los términos (sin importar mayúsculas/minúsculas)
for elemento in cantidad:
    encontrado = False
    for termino in terminos:
        if termino.lower() in os.path.basename(elemento).lower():  # Comparar solo el nombre del archivo
            agrupaciones[termino].append(elemento)
            encontrado = True
            break
    if not encontrado:
        agrupaciones["otros"].append(elemento)

# Patrón para extraer el nombre del incendio
pattern = r'^(.*)_[0-9]{4}'

# Crear las carpetas de destino
path_destino_train_images = r"C:\nico\wildfire2024\data\PyroVideos\images\train"
path_destino_train_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\train"
path_destino_test_images = r"C:\nico\wildfire2024\data\PyroVideos\images\test"
path_destino_test_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\test"
path_destino_val_images = r"C:\nico\wildfire2024\data\PyroVideos\images\val"
path_destino_val_labels = r"C:\nico\wildfire2024\data\PyroVideos\labels\val"

os.makedirs(path_destino_train_images, exist_ok=True)
os.makedirs(path_destino_train_labels, exist_ok=True)
os.makedirs(path_destino_test_images, exist_ok=True)
os.makedirs(path_destino_test_labels, exist_ok=True)
os.makedirs(path_destino_val_images, exist_ok=True)
os.makedirs(path_destino_val_labels, exist_ok=True)

# Proporciones para dividir entre train, val y test
split_train = 0.8
split_val = 0.1
split_test = 0.1

# Función para copiar archivos a la carpeta correspondiente y renombrarlos
def copiar_archivos(frames, destino_images, destino_labels, nombre_incendio):
    for i, frame in enumerate(frames):
        if frame.endswith('.jpg'):
            # Generar nombres nuevos para las imágenes y etiquetas
            nuevo_nombre_frame = f"{nombre_incendio}_frame{i+1}.jpg"
            nuevo_nombre_label = f"{nombre_incendio}_frame{i+1}.txt"

            # Construir las rutas de destino
            frame_dest = os.path.join(destino_images, nuevo_nombre_frame)
            label = frame.replace('.jpg', '.txt').replace('images', 'labels')
            label_dest = os.path.join(destino_labels, nuevo_nombre_label)

            # Copiar imagen
            if os.path.exists(frame):
                shutil.copy(frame, frame_dest)

            # Copiar etiqueta
            if os.path.exists(label):
                shutil.copy(label, label_dest)
            else:
                # Crear un archivo vacío si no existe la etiqueta
                with open(label_dest, 'w') as f:
                    pass

# Procesar cada clave del diccionario agrupaciones excepto "synthetic"
for key, archivos in agrupaciones.items():
    if key == "synthetic" or key == "hpwren" or key == "random":
        continue  # Saltar los synthetic por ahora
    print(f"Procesando {key} con {len(archivos)} archivos...")

    # Agrupaciones de incendios por clave
    agrupaciones_incendios = defaultdict(list)
    
    for archivo in archivos:
        match = re.match(pattern, os.path.basename(archivo))
        if match:
            nombre_incendio = match.group(1)
            agrupaciones_incendios[nombre_incendio].append(archivo)

    # Separar los incendios con más de 5 frames y los que tienen 5 o menos
    incendios_grandes = {incendio: archivos_incendio for incendio, archivos_incendio in agrupaciones_incendios.items() if len(archivos_incendio) > 5}
    incendios_pequenos = {incendio: archivos_incendio for incendio, archivos_incendio in agrupaciones_incendios.items() if len(archivos_incendio) <= 5}

    # Procesar los incendios grandes (más de 5 frames) y dividirlos en train, val y test
    incendios = list(incendios_grandes.keys())
    random.shuffle(incendios)

    n_train = int(len(incendios) * split_train)
    n_val = int(len(incendios) * split_val)
    n_test = len(incendios) - n_train - n_val

    total_images_train, total_images_val, total_images_test = 0, 0, 0

    # Asignar incendios grandes a train, val y test
    for idx, incendio in enumerate(incendios):
        archivos_incendio = incendios_grandes[incendio]
        # que sean jpg
        archivos_incendio_imagenes = [archivo for archivo in archivos_incendio if archivo.endswith('.jpg')]
        
        if idx < n_train:
            copiar_archivos(archivos_incendio, path_destino_train_images, path_destino_train_labels, incendio)
            total_images_train += len(archivos_incendio_imagenes)
        elif idx < n_train + n_val:
            copiar_archivos(archivos_incendio, path_destino_val_images, path_destino_val_labels, incendio)
            total_images_val += len(archivos_incendio_imagenes)
        else:
            copiar_archivos(archivos_incendio, path_destino_test_images, path_destino_test_labels, incendio)
            total_images_test += len(archivos_incendio_imagenes)

    # Procesar los incendios pequeños (5 o menos frames) y colocarlos solo en train
    total_images_small = 0
    for incendio, archivos_incendio in incendios_pequenos.items():
        archivos_incendio_imagenes = [archivo for archivo in archivos_incendio if archivo.endswith('.jpg')]
        copiar_archivos(archivos_incendio, path_destino_train_images, path_destino_train_labels, incendio)
        total_images_small += len(archivos_incendio_imagenes)

    # Imprimir resultados por categoría
    print(f"Cantidad de incendios grandes (>5 frames) en {key}: {len(incendios_grandes)}")
    print(f"Cantidad de incendios pequeños (<=5 frames) en {key}: {len(incendios_pequenos)}")
    print(f"Total imágenes entrenamiento: {total_images_train + total_images_small}")
    print(f"Total imágenes validación: {total_images_val}")
    print(f"Total imágenes prueba: {total_images_test}")

# Al final, añadir los synthetic al conjunto de train (sin agrupación)
print(f"Procesando synthetic con {len(agrupaciones['synthetic'])} imágenes...")

# Copiar todos los synthetic directamente a train
total_synthetic_images = 0
for archivo in agrupaciones["synthetic"]:
    frame_dest = os.path.join(path_destino_train_images, os.path.basename(archivo))
    label = archivo.replace('.jpg', '.txt').replace('images', 'labels')
    label_dest = os.path.join(path_destino_train_labels, os.path.basename(label))

    # Copiar imagen synthetic a train
    if os.path.exists(archivo):
        shutil.copy(archivo, frame_dest)
        total_synthetic_images += 1

    # Copiar etiqueta correspondiente (si existe)
    if os.path.exists(label):
        shutil.copy(label, label_dest)

print(f"Cantidad de imágenes synthetic procesadas: {total_synthetic_images}")
```
