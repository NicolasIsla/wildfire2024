```python
from PIL import Image
import imagehash
import os

# Ruta de las imágenes
image_dir = r"C:\Users\nicoi\Downloads\Nemo_yolo\Nemo_yolo\DS\images\train"

# Diccionario para almacenar los hashes
hashes = {}

# Recorre el directorio de imágenes y calcula el hash
for image_file in os.listdir(image_dir):
    if image_file.endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif')):
        image_path = os.path.join(image_dir, image_file)
        image = Image.open(image_path)
        
        hash_value = imagehash.phash(image)  
        
        hashes[image_file] = hash_value

threshold = 5  # Definir un umbral para determinar similitud
similar_images = []

# Comparar los hashes entre sí
for img1, hash1 in hashes.items():
    for img2, hash2 in hashes.items():
        if img1 != img2:
            if hash1 - hash2 < threshold:  # Si la diferencia de hash es menor al umbral, son similares
                similar_images.append((img1, img2))

# Mostrar resultados
for img1, img2 in similar_images:
    print(f"Imágenes similares: {img1} y {img2}")
```

```python
from PIL import Image
import imagehash
import os

train_dir = r"C:\Users\nicoi\Downloads\Nemo_yolo\Nemo_yolo\DS\images\train"
val_dir = r"C:\Users\nicoi\Downloads\Nemo_yolo\Nemo_yolo\DS\images\val"
test_dir = r"C:\Users\nicoi\Downloads\Nemo_yolo\Nemo_yolo\DS\images\test"

# Lista para juntar todas las rutas
image_dirs = [train_dir, val_dir, test_dir]

# Diccionario para almacenar los hashes
hashes = {}

# Diccionario para almacenar las secuencias de video
videos_dict = {}

# Set para almacenar imágenes ya agrupadas
used_images = set()

# Recorre los tres directorios y calcula los hashes
for image_dir in image_dirs:
    for image_file in os.listdir(image_dir):
        if image_file.endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif')):
            image_path = os.path.join(image_dir, image_file)
            image = Image.open(image_path)
            
            # Calcula el hash de la imagen (usa average_hash por defecto con hash_size=12)
            hash_value = imagehash.average_hash(image, hash_size=12)
            
            # Almacena el hash en el diccionario, usando el path completo como clave
            hashes[image_path] = hash_value

```

```python
threshold = 20  # Definir un umbral para determinar similitud
video_count = 1  # Contador para nombrar los videos (video1, video2, etc.)

# Agrupar imágenes similares en secuencias de video sin repeticiones
for img1, hash1 in hashes.items():
    if img1 not in used_images:  # Solo proceder si la imagen no ha sido usada en otra secuencia
        video_key = f"video{video_count}"  # Crear un nuevo video
        videos_dict[video_key] = [img1]  # Iniciar la secuencia con img1
        used_images.add(img1)  # Marcar img1 como usada
        
        for img2, hash2 in hashes.items():
            if img1 != img2 and img2 not in used_images:
                if hash1 - hash2 < threshold:  # Si las imágenes son similares
                    videos_dict[video_key].append(img2)  # Agregar img2 a la secuencia
                    used_images.add(img2)  # Marcar img2 como usada
        
        video_count += 1  # Incrementar el contador de videos para la siguiente secuencia

# Mostrar las secuencias de video formadas
for video, frames in videos_dict.items():
    print(f"{video}: {', '.join(frames)}")
```
