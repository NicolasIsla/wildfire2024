```python
import shutil
import os
import random

path = r"C:\Users\nicoi\Downloads\FIGLIB_ANNOTATED_RESIZED\FIGLIB_ANNOTATED_RESIZED"
```

```python
def select_images_from_labels(path_labels, num_empty=1, num_non_empty=3):
    empty_files = []
    non_empty_files = []

    # Revisar cada archivo en la carpeta de etiquetas
    for filename in sorted(os.listdir(path_labels)):  # Ordena alfabéticamente
        if filename.endswith('.txt'):
            filepath = os.path.join(path_labels, filename)
            with open(filepath, 'r') as file:
                contents = file.read().strip()
                image_path = os.path.join(path_labels.replace("labels", ""), filename.replace('.txt', '.jpg'))
                if contents == '':
                    empty_files.append((image_path, filepath))  # Guardar el path de imagen y etiqueta
                else:
                    non_empty_files.append((image_path, filepath))

    selected_empty = random.sample(empty_files, min(len(empty_files), num_empty)) if len(empty_files) >= num_empty else []

    if non_empty_files:
        first_non_empty = non_empty_files[0]  # Primer archivo no vacío ordenado alfabéticamente
        remaining_non_empty = non_empty_files[1:]  # Resto de archivos no vacíos
        selected_non_empty = [first_non_empty] + random.sample(remaining_non_empty, min(len(remaining_non_empty), num_non_empty-1)) if len(remaining_non_empty) >= num_non_empty-1 else []

        return selected_empty + selected_non_empty
    else:
        return selected_empty

def process_incident_folders(path, num_empty=1, num_non_empty=3):
    results = {}
    for incident_folder in os.listdir(path):
        path_labels = os.path.join(path, incident_folder, "labels")
        if os.path.exists(path_labels):
            selected_images_labels = select_images_from_labels(path_labels, num_empty, num_non_empty)
            if selected_images_labels:
                results[incident_folder] = selected_images_labels
    return results
```

```python
# Directorios de destino
dest_images_dir = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data/DS_08/images"
dest_labels_dir = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data/DS_08/labels"

# Asegúrate de que las carpetas de destino existan
os.makedirs(dest_images_dir, exist_ok=True)
os.makedirs(dest_labels_dir, exist_ok=True)
```

```python
selected_images_per_incident = process_incident_folders(path, num_empty=1, num_non_empty=6)
total = 0
random.seed(42)

# Calcular el número de imágenes para train, val y test
Total_images = sum(len(files) for files in selected_images_per_incident.values())
train = int(Total_images * 0.8)
val = int(Total_images * 0.1)
test = Total_images - train - val
conteo = 0

# Definir rutas
path_train_images = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\images\train"
path_train_labels = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\labels\train"
path_val_images = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\images\val"
path_val_labels = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\labels\val"
path_test_images = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\images\test"
path_test_labels = r"C:\Users\nicoi\OneDrive\Escritorio\wilfire2024\data\DS_08\labels\test"

# Crear carpetas si no existen
os.makedirs(path_train_images, exist_ok=True)
os.makedirs(path_train_labels, exist_ok=True)
os.makedirs(path_val_images, exist_ok=True)
os.makedirs(path_val_labels, exist_ok=True)
os.makedirs(path_test_images, exist_ok=True)
os.makedirs(path_test_labels, exist_ok=True)

# Iterar sobre los incidentes y archivos seleccionados
path_images = path_train_images
path_labels = path_train_labels
for incident, files in selected_images_per_incident.items():
    
    for image_path, label_path in files:
       
        

        # Copiar imagen al directorio de destino de imágenes
        shutil.copy(image_path, os.path.join(path_images, os.path.basename(image_path)))

        # Copiar etiqueta al directorio de destino de etiquetas
        shutil.copy(label_path, os.path.join(path_labels, os.path.basename(label_path)))

        conteo += 1
        total += 1
    if conteo < train:
        path_images = path_train_images
        path_labels = path_train_labels

    elif conteo < train + val:
        path_images = path_val_images
        path_labels = path_val_labels
    else:
        path_images = path_test_images
        path_labels = path_test_labels

print(f"Total de imágenes seleccionadas: {total}")
```
