```python
# Función para extraer el ID de Google Drive desde la URL
def extract_drive_id(url):
    return url.split('/d/')[1].split('/view')[0]

# Descargar cada archivo usando gdown
for filename, url in env_links.items():
    if url:  # Verificar que la URL no esté vacía
        drive_id = extract_drive_id(url)
        gdown.download(f"https://drive.google.com/uc?id={drive_id}", filename, quiet=False)
    else:
        print(f"No URL found for {filename}")
```

```python
def contar_archivos_vacios(directorio):
    """
    Cuenta cuántos archivos .txt en el directorio especificado están vacíos.
    
    :param directorio: str, la ruta al directorio donde se encuentran los archivos .txt
    :return: int, la cantidad de archivos .txt vacíos
    """
    contador_vacios = 0
    contador_total = 0
    # Lista todos los archivos en el directorio
    for archivo in os.listdir(directorio):
        # Comprueba si el archivo es un .txt
        if archivo.endswith(".txt"):
            ruta_completa = os.path.join(directorio, archivo)
            # Comprueba si el archivo está vacío
            if os.path.getsize(ruta_completa) == 0:
                contador_vacios += 1
            contador_total += 1

    return contador_vacios, contador_total
```

```python
def explorador(path_images, path_labels):
    images = os.listdir(path_images)
    labels = os.listdir(path_labels)
    print(f"Images: {len(images)} Labels: {len(labels)}")
    print()
    cat = [img.split('_')[0] for img in images]
    

    # contar las categorias
    unique, counts = np.unique(cat, return_counts=True)
    print(f"Videos: {len(unique)}")
    print()
    # mostrar las categorias
    for i in range(len(unique)):
        print(f"{unique[i]}: {counts[i]}")
    
    return unique
```
